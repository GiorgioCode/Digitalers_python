# Conjuntos
---

En Python, tanto los conjuntos (sets) como los diccionarios (dictionaries) son tipos de estructuras de datos que se utilizan para almacenar colecciones de elementos. Sin embargo, existen diferencias importantes entre ellos en términos de cómo se almacenan y acceden a los datos.

Un conjunto (set) en Python es una colección desordenada de elementos únicos. Esto significa que no puede haber duplicados en un conjunto y el orden de los elementos no se mantiene. Los conjuntos se definen utilizando llaves ({}) y los elementos se separan por comas.

Los conjuntos son útiles cuando necesitas almacenar una colección de elementos sin importar el orden y sin duplicados. Puedes realizar operaciones de conjunto como unión, intersección y diferencia de conjuntos utilizando métodos específicos como union(), intersection() y difference().
